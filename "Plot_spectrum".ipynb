```python
import matplotlib.pyplot as plt
import numpy as np

fig,ax = plt.subplots()
x = np.linspace(-1,10,101)

for i in range(100):
  plt.plot(x,(x*i) / 100, color=[0 + i / 100,0,0 + i / 100])

ax.set_title('just a spectrum')
plt.show()
```
